```python
import pandas as pd

def load_csvs(file_paths):
    dataframes = []
    for path in file_paths:
        try:
            df = pd.read_csv(path)
            dataframes.append(df)
            print(f"Loaded {path} successfully.")
        except FileNotFoundError:
            print(f"File {path} not found.")
        except Exception as e:
            print(f"An error occurred while loading {path}: {e}")
    return dataframes

# if __name__ == "__main__":
#     # List of file paths
#     file_paths = [r"C:\Users\paude\Documents\00. Backup 2023\EXTERIOR\MASTER\MA IAAC\03. Cursada\01. Modulo 03\S01 DE\Class Project\DE_Team\Scraped_csvs\data_cost.csv",
#                   r"C:\Users\paude\Documents\00. Backup 2023\EXTERIOR\MASTER\MA IAAC\03. Cursada\01. Modulo 03\S01 DE\Class Project\DE_Team\Scraped_csvs\data_prop_price.csv",
#                 r"C:\Users\paude\Documents\00. Backup 2023\EXTERIOR\MASTER\MA IAAC\03. Cursada\01. Modulo 03\S01 DE\Class Project\DE_Team\Scraped_csvs\data_qual.csv"
#                 ]


if __name__ == "__main__":
    # List of file paths
    file_paths = ["/content/drive/MyDrive/Colab Notebooks/DE/Project/Scraped_csvs/data_cost.csv",
                  "/content/drive/MyDrive/Colab Notebooks/DE/Project/Scraped_csvs/data_prop_price.csv",
                "/content/drive/MyDrive/Colab Notebooks/DE/Project/Scraped_csvs/data_qual.csv"
                ]


    # Load CSV files into dataframes
    dfs = load_csvs(file_paths)
```

```python
merged_df = pd.merge(dfs[0], dfs[1], on='City', how='outer')
merged_df = pd.merge(merged_df, dfs[2], on='City', how='outer')
```

```python
# prompt: create a new column in merged_df from the last string after the comma

merged_df['Country'] = merged_df['City'].apply(lambda x: x.split(', ')[-1])


merged_df.head()
```

```python
# prompt: for numeric data only, replace the NaN's with the average values of the same columns who share a similar Country, change the data frame name to merged_filled1_ df, for numeric data only round the floats to one decimal number

# Create a copy of the dataframe
merged_filled1_df = merged_df.copy()

# Select numeric columns
numeric_columns = merged_filled1_df.select_dtypes(include=['float64', 'int64']).columns

# Group the dataframe by country and calculate the mean for each numeric column
country_means = merged_filled1_df.groupby('Country')[numeric_columns].mean().round(1)

# Replace NaN values in the numeric columns with the corresponding country mean
for col in numeric_columns:
    for country in country_means.index:
        merged_filled1_df.loc[(merged_filled1_df['Country'] == country) & (merged_filled1_df[col].isna()), col] = country_means.loc[country, col]

# Display the first few rows of the dataframe
merged_filled1_df.head()
```

```python
# prompt: check merged_filled1_df for NaN's

merged_filled1_df.isnull().sum()
```

```python
# prompt: for numeric data only, replace the NaN's with the average values of the same columns, change the data frame name to merged_filled2_ df

# Create a copy of the dataframe
merged_filled2_df = merged_filled1_df.copy()

# Select numeric columns
numeric_columns = merged_filled2_df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the average values for each numeric column
column_means = merged_filled2_df[numeric_columns].mean()

# Replace NaN values with the average values
merged_filled2_df[numeric_columns] = merged_filled2_df[numeric_columns].fillna(column_means)

# Display the first few rows of the dataframe
merged_filled2_df.head()
```

```python
merged_filled2_df.isnull().sum()
```

```python
# prompt: for numeric data only, round the data in all the merged_filled2_df to one decimal figure, then round the data in Rank_x and Rank_y to integers

import numpy as np

# Round numeric data to one decimal figure
for col in merged_filled2_df.select_dtypes(include=['float64', 'int64']):
    merged_filled2_df[col] = np.round(merged_filled2_df[col], 1)

# Round Rank_x and Rank_y to integers
merged_filled2_df['Rank_x'] = merged_filled2_df['Rank_x'].astype(int)
merged_filled2_df['Rank_y'] = merged_filled2_df['Rank_y'].astype(int)

# Display the first few rows of the dataframe
merged_filled2_df.head()
```

```python
# prompt: sort merged_filled2_df by Rank_x value, and rename the dataframe as merged_filled3_df

merged_filled3_df = merged_filled2_df.sort_values(by=['Rank_x'])

merged_filled3_df.tail(50)
```
