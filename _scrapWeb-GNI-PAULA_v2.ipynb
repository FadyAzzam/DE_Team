```python
# Option 2 (trough URL --> recommended)

# Define url to scrape
country_code = 'COL'
url_string = 'https://databank.worldbank.org/views/reports/reportwidget.aspx?Report_Name=CountryProfile&Id=b450fd57&tbar=y&dd=y&inf=n&zm=n&country=' + country_code

driver.get(url_string)
```

```python
# Find Table element
# hdr = driver.find_element(By.ID,'container-header') #--> Header
tbl = driver.find_element(By.ID,'container-data') #--> Data

# # Parse header element into pandas df
# table_html = hdr.get_attribute('outerHTML')
# soup = BeautifulSoup(table_html, 'html.parser')
# # Find all table rows in the parsed HTML
# table_header = soup.find_all('tr')
# # Find all table rows in the parsed HTML
# table_rows = soup.find_all('tr')
# header = []
# for row in table_rows:
#     row_data = []
#     cells = row.find_all(['td', 'th'])
#     for cell in cells:
#         row_data.append(cell.get_text().strip())
#     header.append(row_data)


# Parse Table element into pandas df
table_html = tbl.get_attribute('outerHTML')
soup = BeautifulSoup(table_html, 'html.parser')
# Find all table rows in the parsed HTML
table_rows = soup.find_all('tr')
data = []
for row in table_rows:
    row_data = []
    cells = row.find_all(['td', 'th'])
    for cell in cells:
        row_data.append(cell.get_text().strip())
    data.append(row_data)

# Create DataFrame & append to existing one
df_aux = pd.DataFrame(data[2:], columns=df_countries.columns[0:5])  # Assuming first row is header
df_aux['country_code'] = country_code
df_countries = pd.concat([df_countries, df_aux], ignore_index=True)
```
