```python
url_crime ='https://www.numbeo.com/crime/rankings_current.jsp'
```

```python
df_capital = pd.DataFrame()
```

```python
# Fetch data using Selenium
driver.get(url_crime)

# Find Table element
tbl_crime = driver.find_element(By.XPATH, '//*[@id="t2_wrapper"]')  # if there's no ID, always a good idea to copy xpath

# Parse Table element into pandas DataFrame
table_crime_html = tbl_crime.get_attribute('outerHTML')
soup = BeautifulSoup(table_crime_html, 'html.parser')

# Find all table rows in the parsed HTML
table_crime_rows = soup.find_all('tr')
```

```python
# Initialize data list
data_crime = []
for row in table_crime_rows:
        row_data = []
        cells = row.find_all(['td', 'th'])
        for cell in cells:
            row_data.append(cell.get_text().strip())
        data_crime.append(row_data)

# Print out data structure for debugging
print("Columns:", data_crime[0])
print("Data:", data_crime[1:])
```

```python
# Extract header and data rows separately
header = data_crime[0]
data = data_crime[1:]

# Create DataFrame
df_capital = pd.DataFrame(data, columns=header)

# Display DataFrame
print(df_capital)
df_capital.head(20)
```

```python
# Filter out only the capital cities
capital_cities = []
for city in df_capital['City']:
    if city.split(',')[0] in capitals:
        capital_cities.append(city)

# Display the filtered list of capital cities
print(capital_cities)
print(len(capital_cities))
```
